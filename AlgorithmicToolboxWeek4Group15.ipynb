#Binary Search

```python
%%writefile binary_search.cpp
#include <bits/stdc++.h>

using namespace std;

int binary_search(const vector<int> &a, int x) {
  int left = 0, right = (int)a.size()- 1,med; 
  while(left <= right){
    med = left + (right-left) / 2;
    if(a[med]==x)
        return med;
    if(x > a[med])
        left = med + 1;
    else
        right = med - 1;    
  }
  return -1;
}

int main() {
  int n;
  cin >> n;
  vector<int> a(n);
  for (size_t i = 0; i < a.size(); i++) {
    cin >> a[i];
  }
  int m;
  cin >> m;
  vector<int> b(m);
  for (int i = 0; i < m; ++i) {
    cin >> b[i];
  }
  for (int i = 0; i < m; ++i) {
    cout << binary_search(a, b[i]) << ' ';
  }
}
```

#Majority Element

```python
%%writefile majority_element.cpp
#include <bits/stdc++.h>

using namespace std;

long long ans(vector<long long> A, int n) {
	map<int, int> mapa;

	for (int i = 0; i < n; i++)
		mapa[A[i]]++;
	for (int i = 0;i < n; i++)
		if (mapa[A[i]] > n / 2)
			return 1;
	return 0;
}
int main() {
	long long n;
	cin >> n;

	vector<long long> a(n);
	for (size_t i = 0; i < n; i++) {
		cin >> a[i];
	}
	cout << ans(a, n) << ' ';

	return 0;
}
```

#Improving Quicksort

```python
%%writefile improving_quicksort.cpp
#include <bits/stdc++.h>

using namespace std;

vector<int> nueva_particion(vector<int> &a, int l, int r)
{
    int x = a[l], peq = l, gr = r, i = l;
    while (i <= gr)
    {
        //particion  < x
        if (a[i] < x)
        {
            swap(a[i], a[peq]);
            peq++;
            i++;
        }
        //particion > x
        else if (a[i] > x)
        {
            swap(a[i], a[gr]);
            gr--;
        }
        //particion == x
        else
        {
            i++;
        }
    }

    return vector<int>{peq, gr};
}

void randomized_quick_sort(vector<int> &a, int l, int r)
{
    if (l >= r)
    {
        return;
    }

    int k = l + rand() % (r - l + 1);
    swap(a[l], a[k]);
    vector<int> m = nueva_particion(a, l, r);
    int m1 = m[0];
    int m2 = m[1];
    randomized_quick_sort(a, l, m1 - 1);
    randomized_quick_sort(a, m2 + 1, r);
}

int main()
{
    int n;
    cin >> n;
    vector<int> a(n);
    for (size_t i = 0; i < a.size(); ++i)
    {
        cin >> a[i];
    }
    randomized_quick_sort(a, 0, a.size() - 1);
    for (size_t i = 0; i < a.size(); ++i)
    {
        cout << a[i] << ' ';
    }
}
```
