#Money Change

```python
%%writefile change.cpp
#include <bits/stdc++.h>
using namespace std;
int vueltas(int m) {
  int n=0;
  n += m / 10;
  m=m % 10;
  n += m / 5;
  m = m % 5;
  n += m;
  return n;
}

int main() {
  int m;
  cin >> m;
  cout << vueltas(m) << '\n';
  return 0;
}
```

#Maximum value of loot

```python
%%writefile fractional_knapsack.cpp
#include <bits/stdc++.h>

using namespace std;

struct Loot{
	int valor, peso;
};
bool cmp(Loot x, Loot y){
	double r1=(double)x.valor/x.peso;
	double r2=(double)y.valor/y.peso;
	return r1>r2;
}
void answer(Loot looteo[], int n, int restrict){
		sort(looteo, looteo+n, cmp);
		double dol=0;
		int we=0;
		for(int i=0;i<n;i++){
			if(looteo[i].peso+we<restrict){
				dol+=looteo[i].valor;
				we+=looteo[i].peso;
			}else{
				int temp_peso=restrict-we;
				dol+=((double)temp_peso/(looteo[i].peso))*(looteo[i].valor);
				break;
			}
		}
		cout.precision(10);
		cout << dol << endl;
	}
int main(){
	//a[n]
	Loot *looteo = new Loot[10000];
	int n, restrict ;
	cin >> n >> restrict;
	for(int i = 0; i  < n; i++){
	    cin >> looteo[i].valor ; 
	    cin >> looteo[i].peso ;
	}
	answer(looteo, n, restrict);
	return 0;
}
```

```python
!g++ fractional_knapsack.cpp -o fractional
!./fractional
```

#Car fueling

```python
%%writefile car_fueling.cpp
#include <bits/stdc++.h>

using namespace std;

int relleno(int d ,int maxi,int n,vector<int> gas){
    int ans = 0;
    int i=0;
    int limitante=maxi;
    int miles=maxi;
    while(limitante < d){
        if(i >= n || gas[i] > limitante){
            return -1;
        }        
            while((i < n-1) && (gas[i+1] <= limitante)){
                i++;
            }
            ans++;
            limitante = gas[i]+miles;
            i++;
    }    
    return ans;
}

int main(){
    int d, maxi , n;
    cin >> d>> maxi >> n;
    vector<int> gas(n);
    for(int i = 0; i < n; i++ )
        cin >> gas[i];
    cout<<(relleno(d, maxi , n , gas));    
    return 0;
}
```
